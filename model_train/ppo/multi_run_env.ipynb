```python
def get_stable_data(store_data_path):
    exists_data = []
    for i in store_data_path:
        if os.path.exists(i+'data'):
            exists_data.append(i+'data')
    size_data = [os.path.getsize(i) for i in exists_data]
    time.sleep(2)
    size_data2 =  [os.path.getsize(i) for i in exists_data]
    finish_data = []
    for i in range(len(size_data)):
        finish_data.append(size_data2[i]-size_data[i])
    stable_data = []
    for i in range(len(finish_data)):
        if finish_data[i] == 0:
            stable_data.append(exists_data[i])
    return stable_data
```

```python
def send_data(stable_data, model_train_path):
    data_name = []
    for i in stable_data:
        data_name.append(model_train_path+str(time.time()))
        os.rename(i,data_name[-1])
        time.sleep(0.1)
    return data_name
```

```python
def copy_model(model_output_path):
    trained_model_path = "F:\TX_Distribution\multi_train_temp_model\model.onnx"
    for i in model_output_path:
        while(True):
            try:
                copyfile(trained_model_path,i)
                break
            except:
                continue
    return True
```
