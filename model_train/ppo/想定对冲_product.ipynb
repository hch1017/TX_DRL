```python
import json
import random
import copy
```

```python
data = json.load(open('F:\TX_Distribution\想定_对冲_8000.json','rb'))
```

```python
sample = data['仿真实例'][0]
```

```python
data['仿真实例'] = [sample]
```

```python
for i in sample:
    print (i)
```

```python
for _ in range(20):
    sample_random = sample.copy()
    for i in sample:
        height = random.randint(4000,8000)
        diff = random.randint(100,3000)
        if i=='红方飞机':
            sample_random[i]['b-高度'] = height
            sample_random[i]['c-方位'] =  random.randint(0,180)
            sample_random[i]['d-俯仰'] =  random.randint(0,10)
            sample_random[i]['e-滚转'] =  random.randint(0,20)
            sample_random[i]['f-速度'] =  random.randint(180,400)
        elif i=='蓝方飞机':
            sample_random[i]['b-与红方横向间隔'] = random.randint(100,2000)
            sample_random[i]['c-与红方纵向间隔'] = diff
            sample_random[i]['d-高度'] = height - diff
            sample_random[i]['e-方位'] =  random.randint(0,180)
            sample_random[i]['f-俯仰'] =  random.randint(0,10)
            sample_random[i]['g-滚转'] =  random.randint(0,20)
            sample_random[i]['h-速度'] =  random.randint(180,400)
    
    print (sample_random)
    new_random = copy.deepcopy(sample_random)
    data['仿真实例'].append(new_random)
```
